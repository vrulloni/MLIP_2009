cv2.arcLength(cnt,True): calcula longitud de arco. Puede ser una curva cerrada o no. Si se le pasa True es un contorno.



```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

src = cv2.imread('herramientas.jpg')

gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)

gray = cv2.GaussianBlur(gray, (7, 7), 3)

t, dst = cv2.threshold(gray, 0, 255, cv2.THRESH_TRIANGLE)

# obtener los contornos
_, contours, _ = cv2.findContours(dst, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

nueva=cv2.drawContours(src, contours, -1, (0, 0,255), 2)

print('Hay ', len(contours),' totales')

plt.imshow(nueva),plt.title('Todos los contornos')
plt.show()

#cv2.LINE_AA es un tipo de lÃ­nea recomendado
suma=0
for c in contours:
    area = cv2.contourArea(c)
    if area > 100 and area < 60000:
        cadauna=cv2.drawContours(src, [c], 0, (0, 0,255), 2, cv2.LINE_AA)
        suma=suma+1

plt.imshow(cadauna),plt.title('Solo los contornos deseados')
plt.show()

print('Hay',suma,'objetos')
```
